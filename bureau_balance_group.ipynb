```python
import pandas as pd
```

```python
df=pd.read_csv('bureau_balance.csv')
```

```python
df['SK_ID_BUREAU'] = df['SK_ID_BUREAU'].astype(object)
```

```python
numeric_var = [key for key in dict(df.dtypes)
                   if dict(df.dtypes)[key]
                       in ['float64','float32','int32','int64']] # Numeric Variable

cat_var = [key for key in dict(df.dtypes)
             if dict(df.dtypes)[key] in ['object'] ] # Categorical Varible
```

```python
numeric_var
```

```python
cat_var
```

```python
# one-hot encoding of categorical variables
cat_df = pd.get_dummies(df['STATUS'])
```

```python
cat_df['SK_ID_BUREAU']=df['SK_ID_BUREAU']
```

```python
cat_group=cat_df.groupby(['SK_ID_BUREAU'],as_index=False).agg(['mean','sum'])
```

```python
cat_group
```

```python
cat_columns = []

# Iterate through the variables names
for var in cat_group.columns.levels[0]:
    # Skip the id name
    if var != 'SK_ID_BUREAU':
        
        # Iterate through the stat names
        for stat in cat_group.columns.levels[1]:
            # Make a new column name for the variable and stat
            cat_columns.append('status_%s_%s' % (var, stat))
```

```python
cat_group.columns=cat_columns
```

```python
cat_group.reset_index()
```

```python
df.columns
```

```python
num_group=df.loc[:,['SK_ID_BUREAU','MONTHS_BALANCE']].groupby(['SK_ID_BUREAU'],as_index=False).agg(['mean','min','max'])
```

```python
num_columns = []

# Iterate through the variables names
for var in num_group.columns.levels[0]:
    # Skip the id name
    if var != 'SK_ID_BUREAU':
        
        # Iterate through the stat names
        for stat in num_group.columns.levels[1]:
            # Make a new column name for the variable and stat
            num_columns.append('%s_%s' % (var, stat))
```

```python
num_group.columns=num_columns
```

```python
df_group = pd.concat([cat_group, num_group], axis=1, sort=False)
```

```python
df_group.reset_index
```

```python
df_group.reset_index()
```

```python
df_group.to_csv('bureau_balance_grouped2.csv')
```
